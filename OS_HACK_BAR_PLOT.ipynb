```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
os_hack = pd.read_csv('/content/drive/MyDrive/Os _ hackathon/OSHack_BankCustomers.csv')
```

```python
#Bar Plot B1
os_count = os_hack['OperatingSystems'].value_counts().sort_index()
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=os_count.index, y=os_count.values, palette="Set2")

plt.xlabel("Different OS")
plt.ylabel("Count")
plt.title("OS Number Distribution")

for i, count in enumerate(os_count):
    ax.text(i, count + 50, count, ha="center", va="bottom", fontsize=12)

plt.show()
```

```python
#Bar Plot B2
marital_counts = os_hack['marital'].value_counts()
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=marital_counts.index, y=marital_counts.values, palette="Set2")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.title("Marital Status Distribution")

for i, count in enumerate(marital_counts):
    plt.text(i, count + 50, count, ha="center", va="bottom", fontsize=12)

plt.legend(["Count"], loc="upper right")

plt.show()
```

```python
#Bar Plot B3
import matplotlib.pyplot as plt
import seaborn as sns

month_counts = os_hack['month'].value_counts()
month_order = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
month_counts = month_counts.reindex(month_order, fill_value=0)

sns.set(style="darkgrid")

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=month_counts.index, y=month_counts.values, palette="Set3")

plt.xlabel("Month")
plt.ylabel("Count")
plt.title("Clients Contacted in Each Month")

for i, count in enumerate(month_counts):
    ax.text(i, count + 20, count, ha="center", va="bottom", fontsize=12)

plt.show()
```

```python
#Bar Plot B4
import matplotlib.pyplot as plt


grouped_data = os_hack.groupby(['marital', 'housing', 'loan']).size().unstack().fillna(0)

plt.style.use('seaborn-darkgrid')

plt.figure(figsize=(12, 6))
bottom = None

for loan_type in ['yes', 'no']:
    index = [' - '.join(map(str, idx)) for idx in grouped_data.index]

    plt.bar(
        index,
        grouped_data[loan_type],
        label=f'Loan: {loan_type}',
        bottom=bottom
    )
    if bottom is None:
        bottom = grouped_data[loan_type]
    else:
        bottom += grouped_data[loan_type]

plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.title("Distribution of Loans by Marital Status")

plt.legend(title="Loan Status")

plt.xticks(rotation=45)
plt.show()
```

```python
#Bar Plot B5
import matplotlib.pyplot as plt

region_counts = os_hack['Region'].value_counts()

plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar', color='salmon')

plt.xlabel("Region")
plt.ylabel("Count")
plt.title("Region-wise Distribution of Clients")

plt.xticks(rotation=0)

plt.show()
```
