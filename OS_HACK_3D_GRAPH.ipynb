```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
os_hack = pd.read_csv('/content/drive/MyDrive/Os _ hackathon/OSHack_BankCustomers.csv')
```

```python
#3D GRAPH M1

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize = (12,6))
ax = fig.add_subplot(111, projection='3d')


ax.scatter(os_hack['balance'], os_hack['age'], os_hack['duration'], c='b', marker='o', label='Data Points')


ax.set_xlabel('Balance')
ax.set_ylabel('Age')
ax.set_zlabel('Duration')
ax.set_title('3D Scatter Plot: Balance vs. Age vs. Duration')

plt.show()
```

```python
#3D GRAPH M2

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


duration = os_hack['duration']
age = os_hack['age']
campaign = os_hack['campaign']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_trisurf(duration, age, campaign, cmap='viridis')

ax.set_xlabel('Contact Duration')
ax.set_ylabel('Age')
ax.set_zlabel('Number of Campaigns')

fig.colorbar(surface, shrink=0.5, aspect=5)

ax.set_title('3D Surface Plot: Contact Duration vs. Age vs. Campaign')

plt.show()
```

```python
#3D GRAPH M3

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


duration = os_hack['duration']
months = os_hack['month']
campaign = os_hack['campaign']

month_mapping = {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
                 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12}
months_numeric = os_hack['month'].map(month_mapping)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(months_numeric, campaign, duration, label='Contact Duration')

ax.set_xlabel('Month')
ax.set_ylabel('Campaign')
ax.set_zlabel('Contact Duration')

ax.set_title('3D Line Plot: Contact Duration vs. Month vs. Campaign')

plt.legend()
plt.show()
```

```python
#3D GRAPH M4

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

browser = os_hack['Browser']
campaign = os_hack['campaign']
duration = os_hack['duration']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(browser, campaign, duration, cmap='viridis')

ax.set_xlabel('Browser')
ax.set_ylabel('Campaign')
ax.set_zlabel('Contact Duration')

ax.set_title('3D Surface Plot: Browser vs. Campaign vs. Contact Duration')

plt.show()
```

```python
# 3D GRAPH M5

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd


df = os_hack[['age', 'Region', 'balance']]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(df['age'], df['Region'], df['balance'], c=df['balance'], cmap='viridis', s=50)

ax.set_xlabel('Age')
ax.set_ylabel('Region')
ax.set_zlabel('Balance')
ax.set_title('3D Scatter Plot: Age vs. Region vs. Balance')

cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Balance')

plt.show()
```
