```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
os_hack = pd.read_csv('/content/drive/MyDrive/Os _ hackathon/OSHack_BankCustomers.csv')
```

```python
#Histogram H1

sns.set(style = "whitegrid")
plt.figure(figsize=(10,6))
sns.histplot(data = os_hack , x = 'age' , bins = 20 , kde = True , palette = "Set2")
plt.xlabel('Age')
plt.ylabel('Count')
plt.title("Age Distribution")
plt.savefig('graph1.png')
```

```python
#Histogram H2

sns.set(style = "whitegrid")
plt.figure(figsize = (10,6))
sns.histplot(data = os_hack , x = 'day' , kde = True )
plt.xlabel('Day')
plt.ylabel('Count')
plt.title('Age Distribution')
```

```python
#Histogram H3

import matplotlib.pyplot as plt


age_data = os_hack['age']

plt.figure(figsize=(10, 6))
plt.hist(age_data, bins=20, color='skyblue', edgecolor='black')

plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Clients")

plt.show()
```

```python
#Histogram H4

import matplotlib.pyplot as plt


secondary = os_hack[os_hack['education'] == 'secondary']
tertiary = os_hack[os_hack['education'] == 'tertiary']
primary = os_hack[os_hack['education'] == 'primary']


plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(secondary['balance'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Balance")
plt.ylabel("Count")
plt.title("Income Distribution (Secondary Education)")


plt.subplot(1, 3, 2)
plt.hist(tertiary['balance'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel("Balance")
plt.ylabel("Count")
plt.title("Income Distribution (Tertiary Education)")

plt.subplot(1, 3, 3)
plt.hist(primary['balance'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel("Balance")
plt.ylabel("Count")
plt.title("Income Distribution (Primary Education)")

plt.tight_layout()

plt.show()
```
