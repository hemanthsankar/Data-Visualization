```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
os_hack = pd.read_csv('/content/drive/MyDrive/Os _ hackathon/OSHack_BankCustomers.csv')
```

```python
#Pie Chart P1
import matplotlib.pyplot as plt
import seaborn as sns


education_counts = os_hack['education'].value_counts()
labels = education_counts.index
sizes = education_counts.values

sns.set(style="whitegrid")

plt.figure(figsize=(8, 8))
sns.color_palette("pastel")

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Education Level Distribution")

plt.axis('equal')
plt.show()
```

```python
#Pie Chart P2
import matplotlib.pyplot as plt

default_counts = os_hack['default'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(default_counts, labels=default_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightgreen'])

plt.title("Default Status Distribution")

plt.axis('equal')
plt.show()
```

```python
#Pie Chart P3
import matplotlib.pyplot as plt

housing_loan_counts = os_hack['housing'].value_counts()
personal_loan_counts = os_hack['loan'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].pie(housing_loan_counts, labels=housing_loan_counts.index, autopct='%1.1f%%', startangle=140)
axes[0].set_title("Housing Loan Status")

axes[1].pie(personal_loan_counts, labels=personal_loan_counts.index, autopct='%1.1f%%', startangle=140)
axes[1].set_title("Personal Loan Status")

plt.show()
```

```python
#Pie Chart P4
import matplotlib.pyplot as plt

grouped_data = os_hack.groupby(['education', 'default']).size().unstack().fillna(0)

education_levels = grouped_data.index
default_yes_counts = grouped_data['yes']
default_no_counts = grouped_data['no']

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].pie(default_yes_counts, labels=education_levels, autopct='%1.1f%%', startangle=140)
axes[0].set_title("Default 'Yes' Distribution by Education")

axes[1].pie(default_no_counts, labels=education_levels, autopct='%1.1f%%', startangle=140)
axes[1].set_title("Default 'No' Distribution by Education")

fig.suptitle("Distribution of Default Status by Education Level")

plt.show()
```

```python
#Pie Chart P5
import matplotlib.pyplot as plt


grouped_data = os_hack.groupby(['housing', 'default']).size().unstack().fillna(0)

housing_labels = grouped_data.index
default_yes_counts = grouped_data['yes']
default_no_counts = grouped_data['no']

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].pie(default_yes_counts, labels=housing_labels, autopct='%1.1f%%', startangle=140)
axes[0].set_title("Default 'Yes' for Clients with Housing Loans")

axes[1].pie(default_no_counts, labels=housing_labels, autopct='%1.1f%%', startangle=140)
axes[1].set_title("Default 'No' for Clients with Housing Loans")

fig.suptitle("Distribution of Default Status for Clients with Housing Loans")

plt.show()
```
